# TF-IDF and N-gram Analysis

```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# Define the texts
texts_tfidf = [
    "I really love the atmosphere of this place.",
    "The service is terrible, I didn't enjoy it.",
    "The food was good, but the service needs improvement.",
    "I love this place, but I won't return because of the service."
]

# TF-IDF calculation process
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts_tfidf)

# Put TF-IDF results in a dataframe
df_tfidf = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Print results
df_tfidf
```

```python
# N-Gram analysis
ngram_vectorizer = CountVectorizer(ngram_range=(2, 3))  # 2-gram ve 3-gram

# N-Gram vectorisation
X_ngram = ngram_vectorizer.fit_transform(texts_tfidf)

# Put N-Gram results in a dataframe
df_ngram = pd.DataFrame(X_ngram.toarray(), columns=ngram_vectorizer.get_feature_names_out())

# Print results
df_ngram
```
